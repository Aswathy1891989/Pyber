```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def reading_file(path):
    #variable path is a string which contains the address of .csv file 
    filePath=open(path)
    df=pd.read_csv(filePath)
    #name=filePath.split("/")[1]
    #df_name=
    #return the dataframe
    return df
```

```python
city_df=reading_file("raw_data/city_data.csv")
city_df.head()
```

```python
ride_df=reading_file("raw_data/ride_data.csv")
```

```python
print("City COunt",len(city_df["city"].unique()))
print("Ride City COunt",len(ride_df["city"].unique()))
```

```python
city_ride_df=city_df.merge(ride_df,on="city",how="outer").sort_values("city")
#city_ride_df=city_df.merge(ride_df,on="city",how="outer")
city_ride_df.head()
```

```python
city_table = city_ride_df.pivot_table(index='city', values='fare', aggfunc=np.mean)
city_table.head()
```

```python
ride_table = city_ride_df.pivot_table(index='city', values='driver_count', aggfunc=np.mean)
ride_table.head()
```

```python
def city_ride():
    # get the unique school id
    city_group_df=city_ride_df["city"].unique()
    # create an empty list
    city_list=[]
    # for each ID repeat
    for city in city_group_df:
        # retrieve all info from the merged df based on ID
        city_g_df = city_ride_df[city_ride_df["city"] == city]
        # retreive the required details
        total_ride=city_g_df["ride_id"].count()
        city_type=city_g_df["type"].unique()[0]
        driver_count=city_g_df["driver_count"].unique()[0]        
        avg_fare=city_g_df["fare"].sum()/total_ride
        #append details to list
        city_list.append([city,total_ride,city_type,driver_count,avg_fare])
    #create df using list and return    
    city_tot=pd.DataFrame(city_list,columns=['City','Total Ride','Type','Driver Count','Fare Per City'])
    return city_tot
```

```python
city_sum=city_ride()
city_sum.head()
```

```python
#s=[10+city_sum['Driver Count']*0.01]
test = [i **0.5*100  for i in city_sum["Driver Count"]]
#sns.factorplot(x="Total Ride", y="Fare Per City", hue="Type",size=5,data=city_sum, kind="swarm",scatter_kws={"s": test})

g = sns.lmplot(x="Total Ride", y="Fare Per City", hue="Type", data=city_sum,scatter_kws={"s": test}, palette=dict(Urban="gold", Suburban="lightskyblue",Rural="lightcoral"))
plt.show()
```

```python
test = [i **0.5*100  for i in city_sum["Driver Count"]]
pal = dict(Urban="lightcoral", Suburban="lightskyblue",Rural="gold")
kws = dict(s=test, linewidth=.5, edgecolor="black")
g = sns.FacetGrid(city_sum, hue="Type", palette=pal,
                   hue_order=["Urban", "Suburban","Rural"],legend_out=False,size=10)
g = (g.map(plt.scatter,"Total Ride","Fare Per City", **kws).add_legend())

plt.show()
```

```python
type_table = city_ride_df.pivot_table(index='type', values='fare', aggfunc=np.sum).reset_index()
type_table.head()
```

```python
#Gold, Light Sky Blue, and Light Coral
colors = ["gold","lightskyblue","lightcoral"]
explode = (0,0,0.1)
plt.pie(type_table["fare"], explode=explode, labels=type_table["type"], colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.title("% of Total Fares by City Type")
plt.show()
```
